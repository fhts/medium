# Mudança de nome de colunas

Quando obtemos os dados de uma base, muito dos nomes de colunas são siglas ou códigos. Para facilitar o processo de trabalho e entendimento, podemos trocar o nome das counas, o que pode trazer maior produtividade e facilidade no trabalho com estes dados.

Na base Adult, sabemos que a coluna fnlwgt é um código para cada pessoa pesquisada. Para facilitar o entendimento quando formos trabalhar novamente com essa base, iremos alterar para o nome Id utilizando rename().

__[Documentação rename](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html)__

```python
data1 = data.rename(columns={'fnlwgt': 'Id'})
data1.head()
```

# Tirar coluna

Quando coletamos dados, muitas vezes nós acabamos obtendo muito mais dados que precisavamos. Esses dados extras ocupam espaço na memória e aumentam a dimensionalidade dos dados, sendo interesante retira-los.

Na base Adult, education e educational-num apresentam a mesma informação sobre o nível educacional da pessoa, sendo que a diferença é se este dado é apresentado de forma numérica ou em texto. Para exemplificar esse caso, iremos retira a coluna education com drop().


__[Documentação drop](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html)__

```python
data2 = data.drop(['education'], axis=1)
data2.head()
```
